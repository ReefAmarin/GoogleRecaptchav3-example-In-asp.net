# **Simple line plot**

** Perhaps the simplest of all plots is the visualization of a single function  y=f(x) .As with all the following sections, we'll start by setting up the notebook for plotting and importing the packages we will use:**

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
import numpy as np
```

**we start by creating a figure and an axes**.

```python
fig = plt.figure()
ax = plt.axes()
```

```python
fig = plt.figure()
ax1 = plt.axes()
ax2 = plt.axes()

x = np.linspace(0, 10, 1000)
ax1.plot(x, np.sin(x));
ax2.plot(x, np.cos(x));
```

**Adjusting the Plot: Axes Limits**

```python
plt.plot(x, np.sin(x))

plt.xlim(0, 10)
plt.ylim(-1, 1);
```

The plt.axis() method allows you to set the x and y limits with a single call, by passing a list which specifies [xmin, xmax, ymin, ymax]:

```python
plt.plot(x, np.sin(x))
plt.axis([-1, 11, -1.5, 1.5]);
```

```python
plt.plot(x, np.sin(x))
plt.axis('tight');
```

```python
fig=plt.figure()
ax=plt.axes()
ax.plot(x, np.sin(x))
ax.axis('equal'); 
#ax.xlim(0,4) 
ax.set_xlim(0,4)
```

```python
mycolor_pal=['#FFFF00','#C8A4BA','#C8B4BA','#C8B4BA','#C8B4BA']
```

```python
plt.plot(x, (x - 0),':', color='blue')        # specify color by name
plt.plot(x, (x - 1), '--',color='r')           # short color code (rgbcmyk)
plt.plot(x, (x - 2),'*-', color='0.5')        # Grayscale between 0 and 1
plt.plot(x, (x - 3), color=mycolor_pal[0])     # Hex code (RRGGBB from 00 to FF)
plt.plot(x, (x - 4), color=(0.128,0.128,0)) # RGB tuple, values 0 to 1
plt.plot(x, (x - 5), color='chocolate'); # all HTML color names supported
```

**Labeling Plots**

1-Titles and axis labels 

```python
plt.plot(x, np.sin(x))
plt.title("A Sine Curve")
plt.xlabel("x")
plt.ylabel("sin(x)");
```

For transitioning between MATLAB-style functions and object-oriented methods, make the following changes:

plt.xlabel() → ax.set_xlabel()

plt.ylabel() → ax.set_ylabel()

plt.xlim() → ax.set_xlim()

plt.ylim() → ax.set_ylim()

plt.title() → ax.set_title()

```python
ax=plt.axes()
ax.plot(x, np.sin(x))
ax.set_title("A Sine Curve")
ax.set_xlabel("x")
ax.set_ylabel("sin(x)");
ax.legend(["sin(x)"]);
```

 **Matplotlib Gotchas**
